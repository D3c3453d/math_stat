```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
N = 100
theta = 10
```

```python
T1 = []
T2 = []
T3 = []
T4 = []
T5 = []
for n in range(1, N + 1):
  Xn = []
  for i in range(n):
    Xn.append(np.random.random()*theta)

  T1.append(np.fabs(2 * np.mean(Xn) - theta))
  T2.append(np.fabs((n + 1) * np.min(Xn) - theta))
  T3.append(np.fabs((n + 1)/n * np.max(Xn) - theta))
  T4.append(np.fabs(np.min(Xn) + np.max(Xn) - theta))
  if n > 1:
    T5.append(np.fabs(Xn[0] + np.mean(Xn[1:]) - theta))
  else:
    T5.append(np.fabs(Xn[0] - theta))
```

```python
fig, ax = plt.subplots()
fig.suptitle("θ1")
ax.grid(True)
ax.set_ylabel('θ')
ax.set_xlabel('n')
ax.plot(range(1, N + 1), T1)
```

```python
fig, ax = plt.subplots()
fig.suptitle("θ2")
ax.grid(True)
ax.set_ylabel('θ')
ax.set_xlabel('n')
ax.plot(range(1, N + 1), T2)
```

```python
fig, ax = plt.subplots()
fig.suptitle("θ3")
ax.grid(True)
ax.set_ylabel('θ')
ax.set_xlabel('n')
ax.plot(range(1, N + 1), T3)
```

```python
fig, ax = plt.subplots()
fig.suptitle("θ4")
ax.grid(True)
ax.set_ylabel('θ')
ax.set_xlabel('n')
ax.plot(range(1, N + 1), T4)
```

```python
fig, ax = plt.subplots()
fig.suptitle("θ5")
ax.grid(True)
ax.set_ylabel('θ')
ax.set_xlabel('n')
ax.plot(range(1, N + 1), T5)
```
